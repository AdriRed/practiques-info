## Exercici 2

```python
x = np.linspace(0, 2*np.pi, 100)

y = np.cos(2*x-np.pi)/(x+1)

# help(plt.plot)


plt.plot(x, y, "or", markevery=10)
plt.plot(x, y, "b--")

plt.ylim(-1, 1)
plt.xlim(0, 2*np.pi)

plt.title("Exercici 2")
plt.xlabel("Angle")
plt.ylabel("cos(2X-$\pi$)/(X+1)")

plt.annotate("6è punt", xy=(x[50], y[50]), arrowprops=dict(facecolor= "black"), xytext=(3, -0.75))

plt.show()
```

## Exercici 3

```python
t = np.linspace(0, 2*np.pi, 100)

def lissajous(alpha, beta):
    x = np.sin(alpha*t)
    y = np.sin(beta*t)
    plt.plot(x, y, "r")

plt.subplot(1, 3, 1)
lissajous(4, 3)

plt.subplot(1, 3, 2)
lissajous(6, 3)

plt.subplot(1, 3, 3)
lissajous(2, 5)

plt.show()

```

## Exercici 4

```python
import scipy.optimize as so
import math

def func_quadr(v):
    return (v**2)/5 - 1

x = np.linspace(-10, 10, 100)

ycos = np.cos(x)
ysin = np.sin(x)
ysqr = func_quadr(x)


arrel_cos = so.newton(np.cos, -0.5)
arrel_sin = so.newton(np.sin, -0.5)
arrel_quadr = so.newton(func_quadr, -0.5)


plt.ylim(-1.2, 1.2)
plt.axhline(color="black") #Eix x

plt.plot(x, ysin)
plt.plot(x, ycos)
plt.plot(x, ysqr)

plt.annotate("Arrel $\\cos$", xy=(arrel_cos, 0), arrowprops=dict(facecolor= "black"), xytext=(0, -0.5))
plt.annotate("Arrel $\\sin$", xy=(arrel_sin, 0), arrowprops=dict(facecolor= "black"), xytext=(1.5, 0.5))
plt.annotate("Arrel $\\frac{x^2}{5}-1$", xy=(arrel_quadr, 0), arrowprops=dict(facecolor= "black"), xytext=(-7, 0.5))


plt.show()
```

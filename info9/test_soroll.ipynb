```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from collections.abc import Iterable   # import directly from collections for Python < 3.3

sorolls = set()

data = []

def check_list_type(A: list, t: type):
    """
    Comprova de forma recursiva els tipus de l'array sencer
    """

    return all(check_list_type(inner, t) if (isinstance(inner, Iterable)) # per cada element d'A:
                        else (type(inner) is t)             #   - si es una llista comprova el tipus de cada element de la llista amb la mateixa funcio
                    for inner in A)                         #   - si no es una llista fa la comparacio tal qual per l'element


with open('soroll_bcn.csv', "r") as file:
    csv_reader = csv.DictReader(file)
    for i in csv_reader:
        sorolls.add(i["TOTAL_D"])
        dic = {
            "soroll_diurn": i["TOTAL_D"], 
            "localitzacio": 
                (eval(i["GEOM_WKT"].replace(' ', ',')[16:]))
        }
        if (len(dic["localitzacio"]) % 2 != 0 or any(type(x) is tuple for x in dic["localitzacio"])):
            continue
        dic["localitzacio"] = np.array(dic["localitzacio"]).reshape(-1, 2)
        data.append(dic)
        

sorolls = sorted(list(sorolls))
sorolls = [sorolls[-1]] + sorolls[0:-1]

data = {(coord[0], coord[1]): sorolls.index(d["soroll_diurn"]) for d in data for coord in d["localitzacio"]}

data
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

X, Y = np.meshgrid(x, y)
print(x, y)
plt.imshow(f(X, Y))
```

```python
def func(x, y):
    if ((x, y) in data):
        return data[(x,y)]
    return -1

x = np.array([a[0] for a in data])
y = np.array([a[1] for a in data])

plt.imshow(func(x, y))
```

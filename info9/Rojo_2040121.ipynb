## Exercici 3

````python
k = np.linspace(0, 2*np.pi, 100)

def lissajous(alpha, beta):
    """
    Crea una figura de lissajous amb l'alpha i beta asignades i crea el grafic
    """
    x = np.sin(alpha*k) # x que seràn imatges de k amb la formula i l'alpha asignats
    y = np.sin(beta*k) # igual que adalt pero y i beta
    
    plt.plot(x, y, "r") # creacio del plot

plt.subplot(1, 3, 1, ) # preparo el primer plot
lissajous(4, 3) # creo la figura, es repeteix aixi que no comento mes

plt.subplot(1, 3, 2)
lissajous(6, 3)

plt.subplot(1, 3, 3)
lissajous(2, 5)

# plt.subplots_adjust(wspace=0.45)

plt.show()

````

## Exercici 4

````python
import scipy.optimize as so
import math

def func_quadr(k):
    """
    Funcio quadratica: (x**2)/5 - 1
    """
    return (k**2)/5 - 1

x = np.linspace(-5, 5, 100)

ycos = np.cos(x) # imatges de funcio cos
ysin = np.sin(x) # ...
ysqr = func_quadr(x)


arrel_cos = so.newton(np.cos, -0.5) # trobar arrel propera a -0,5 de la funcio cos
arrel_sin = so.newton(np.sin, -0.5) # ...
arrel_quadr = so.newton(func_quadr, -0.5)


plt.ylim(-1.2, 1.2)
plt.axhline(color="black") #Eix x

plt.plot(x, ysin)
plt.plot(x, ycos)
plt.plot(x, ysqr)

plt.annotate("Arrel $\\cos(x)$", xy=(arrel_cos, 0), arrowprops=dict(facecolor= "black"), xytext=(0, -0.5)) # utilitzo l'arrel per ficar la fletxa
plt.annotate("Arrel $\\sin(x)$", xy=(arrel_sin, 0), arrowprops=dict(facecolor= "black"), xytext=(1.5, 0.5))
plt.annotate("Arrel $\\frac{x^2}{5}-1$", xy=(arrel_quadr, 0), arrowprops=dict(facecolor= "black"), xytext=(-5, 0.5))


plt.show()
````

````python
import csv
import numpy as np
import matplotlib.pyplot as plt
from collections.abc import Iterable   # import directly from collections for Python < 3.3

sorolls = set()

data = []

def check_list_type(A: list, t: type):
    """
    Comprova de forma recursiva els tipus de l'array sencer
    """

    return all(check_list_type(inner, t) if (isinstance(inner, Iterable)) # per cada element d'A:
                        else (type(inner) is t)             #   - si es una llista comprova el tipus de cada element de la llista amb la mateixa funcio
                    for inner in A)                         #   - si no es una llista fa la comparacio tal qual per l'element


with open('soroll_bcn.csv', "r") as file:
    csv_reader = csv.DictReader(file)
    for i in csv_reader:
        sorolls.add(i["TOTAL_D"])
        dic = {
            "soroll_diurn": i["TOTAL_D"], 
            "localitzacio": 
                (eval(i["GEOM_WKT"].replace(' ', ',')[16:]))
        }
        if (len(dic["localitzacio"]) % 2 != 0 or any(type(x) is tuple for x in dic["localitzacio"])):
            continue
        dic["localitzacio"] = np.array(dic["localitzacio"]).reshape(-1, 2)
        data.append(dic)
        

sorolls = sorted(list(sorolls))
sorolls = [sorolls[-1]] + sorolls[0:-1]

data = {(coord[0], coord[1]): sorolls.index(d["soroll_diurn"]) for d in data for coord in d["localitzacio"]}

data
````
